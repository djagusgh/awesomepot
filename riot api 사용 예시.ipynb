### 1. Champion-MASTERY-V4

https://developer.riotgames.com/api-methods/#champion-mastery-v4/GET_getAllChampionMasteries

- 사용자의 사용 챔피언별 정보 보여줌

<br>

#### 메서드 목록

a) getAllChampsInfo_byEncryptedSummonerId(self, encryptedId)

b) getChampInfo_byEncryptedSummonerId_and_championId(self, encryptedId, championId)

c) getChampScores_byEncryptedSummonerId(self, encryptedId)

```python
# 챔피언 코드 json 파일 불러오기
with open('champions.json', encoding="UTF-8") as json_file:
    champsData = json.load(json_file)

# champion key 숫자 순으로 정렬
champsData.sort(key = lambda x : x['key'])
print(champsData[:1])
```

####  a) 유저의 챔피언 정보(챔피언 이름, 숙련도, 점수, 플레이 시간 등등)

```python
c = ChampionMasteryInfo(region, api_key)
# dictionary를 각 원소로 하는 리스트 출력
userChampInfos = c.getAllChampsInfo_byEncryptedSummonerId(encryptedId)
userChampInfos[:2]
```

#### a) -1 예시 : 유저가 플레이한 챔피언 이름들 출력

```python
# 예시 - 유저가 플레이한 챔피언 출력하기

temp = [each['championId'] for each in userChampInfos]
userChamps = []
for a in temp:
    for champ in champsData:
        if int(champ['key']) == a:
            userChamps.append(champ['id'])
# 많이 한 챔프 순으로 출력됨
print(userChamps)
```

#### c) 챔피언 숙련도 총합 출력

```python
c.getChampScores_byEncryptedSummonerId(encryptedId)
```
